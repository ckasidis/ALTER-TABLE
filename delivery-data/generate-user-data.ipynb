```python
import pandas as pd
import numpy as np
import random
```

```python
numberFormat = "+650000"

numbers = np.array([], dtype=str)
orderIDs = np.array([], dtype=np.int32)
addressIDs = np.array([], dtype=np.int32)
parcelNumbers = np.array([], dtype=np.int32)

for i in range(1, 101):

    orderIDs = np.append(orderIDs, [1000 + i * 7]) 

    i = str(i)
    while len(i) < 4:   i = "0" + i

    numbers = np.append(numbers, [numberFormat + i])
    addressIDs = np.append(addressIDs, random.randint(1, 10))
    parcelNumbers = np.append(parcelNumbers, np.floor(np.abs(np.random.normal(0, 1.5)) + 1))

parcelNumbers = parcelNumbers.astype(np.int32)
```

```python
df = pd.DataFrame(
    {"userID": numbers,
    "orderID" : orderIDs,
    "addressID" : addressIDs,
    "parcelNumber" : parcelNumbers}
)

df.head(20)
```
